# 2. Aquirir secuencias de GENBANK
Descargar secuencias de GenBank

```python
accessions = [
    "KX758147.1",
    "KX855639.1",
    "KX855659.1",
    "KX758148.1",
    "KR028300.1",
    "MG273498.1",
    "KR028316.1",
    "KR028311.1",
    "KX758149.1",
    "KR028317.1",
    "KX758146.1",
    "KX758145.1",
    "JQ406623.1",
    "KR028312.1",
    "MG273464.1",
    "MG273497.1"
]

with Entrez.efetch(
    db="nucleotide",
    id=",".join(accessions),
    rettype="gb",
    retmode="text"
) as handle:
    records = list(SeqIO.parse(handle, "genbank"))

print("records existe:", "records" in globals())
print("número de registros:", len(records))
```

Descargar solo las secuencias del fragmento 16S + tRNA-Leu
1. Definir una función
2. Buscar un  feature.qualifiers.get : note, product y gene
3. Buscar 16S


```python
from Bio.SeqRecord import SeqRecord

def extract_16S_flexible(record):
    extracted = []

    for feature in record.features:
        text = " ".join(
            feature.qualifiers.get("note", []) +
            feature.qualifiers.get("product", []) +
            feature.qualifiers.get("gene", [])
        ).lower()

        if "16s" in text:
            try:
                seq = feature.extract(record.seq)
                extracted.append(
                    SeqRecord(
                        seq,
                        id=f"{record.id}|16S",
                        description=""
                    )
                )
            except:
                pass

    return extracted
```

Gruardar todas las secuencias en un multifasta

```python
all_16S = []

for record in records:
    all_16S.extend(extract_16S_flexible(record))

from Bio import SeqIO
SeqIO.write(all_16S, "16S_only.fasta", "fasta")

print("16S extraídas:", len(all_16S))
```
